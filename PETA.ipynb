```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
import os
```

```python
urls = ['https://crueltyfree.peta.org/companies-dont-test/',]
comapny_names = []
links = []
urls = []
i = 1
while i < 125:
    url = f'https://crueltyfree.peta.org/companies-dont-test/page/{i}/'
    requests.get(url).raise_for_status()
    urls.append(url)
    i +=1


```

```python
n = len(urls)

for index, url in enumerate(urls):
    print(f'progress: {index}/{n}', flush = True)
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')
    rows = soup.find_all(rel='bookmark')
    for row in rows:
        comapny_names.append(row.text)
        link = row.attrs['href']
        links.append(link)

print(rows)




```

```python
m= len(links)
folder_path = r'C:\Users\majci\OneDrive\Pulpit\resultsPETA'
errors = []
all_companies = []
n = len(links)
for index, link in enumerate(links):
    print(f'progress: {index}/{n}', flush = True)
    result = requests.get(link)
    soup = BeautifulSoup(result.text, 'html.parser')

    try:
        Company_name = soup.find('span', id ="ctl00_ContentPlaceHolder1_l_CompanyName1" ).text
        comapny_products = soup.find('ul', id="ctl00_ContentPlaceHolder1_dl_Products").text
        compaany_products_list = list(comapny_products.split('\n'))
        compaany_products_list.remove('')
        compaany_products_list.remove('')
        company_information = {
            "CompanyName": Company_name,
            "AnimalTesting": False,
            "CompanyProducts": compaany_products_list
        }


        img = soup.find('div', class_ ='company-logo')
        logo =img.find('img')['src']
        if 'what-is-beauty-without-bunnies-' not in logo:
            company_information["Logo"] = logo

        
        article = soup.find('section', class_='entry-content')


        if soup.find('dl', class_='company-facts-list') in article:
            if "All products are vegan." in article.text:
                company_information["Vegan"] = True
            else:
                company_information["Vegan"] = False

            if "Features PETA’s Global Beauty Without Bunnies Logo" in article.text:
                company_information["FeaturesPetaLogo"] = True
            else:
                company_information["FeaturesPetaLogo"] = False

            
            if 'This company is a PETA Business Friend.' in article.text:
                company_information["BusinessFriend"] = True
            else:
                company_information["BusinessFriend"] = False
        else: 
            company_information["Vegan"] = False
            company_information["FeaturesPetaLogo"] = False
            company_information["BusinessFriend"] = False


        all_companies.append(company_information)
        with open(r'C:\Users\majci\OneDrive\Pulpit\PETA2.json', 'w', encoding='utf-8') as file:   
                json.dump(all_companies, file, ensure_ascii=False, indent=4)
    except Exception as error:
        errors.append((link, error))
```
